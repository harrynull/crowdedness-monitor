```python
import pandas as pd
import numpy as np
import json
```

```python
data = pd.DataFrame(json.load(open('yfWN0nNz.json'))['data'])
```

```python
from datetime import datetime
```

```python
data['time_new'] = data['time'].map(lambda x:datetime.fromtimestamp(x))
```

```python
data['time_new']
```

```python
data = data[data['time_new'] >= '2019-10-25 08:05:26']
```

```python
data
```

```python
data.index = range(len(data))
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(20,6))
ax = fig.add_subplot(111)
ax.plot(data['time_new'], data['universal_mac_count'])
plt.show()
```

```python
max_num = data['universal_mac_count'].max()
avg_num = data['universal_mac_count'].mean()
```

```python
max_num
```

```python
avg_num
```

```python
percent = 0.9
data['crowdness'] = data['universal_mac_count'] / (max_num*percent)
```

```python
data.head()
```

```python
def to_1(x):
    return min(1,x)
data['crowdness'] = data['crowdness'].map(lambda x:to_1(x))
```

```python
data.head()
```

```python
data.head(13)
```

```python
def to_hour(x):
    return x.hour
data['hour'] = data['time_new'].map(lambda x: to_hour(x))
```

```python
data.head(13)
```

```python
def to_mini(x):
    return x.minute
data['mini'] = data['time_new'].map(lambda x: to_mini(x))
```

```python
data.head(13)
```

```python
featureCatCols = ['mini']
dataFeatureCat = data[featureCatCols]
dataFeatureCat = dataFeatureCat.fillna('NA')
X_dictCat = dataFeatureCat.T.to_dict().values()
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn import preprocessing
```

```python
vec = DictVectorizer(sparse = False)
X_vec_cat = vec.fit_transform(X_dictCat)
```

```python
enc = preprocessing.OneHotEncoder()
enc.fit(X_vec_cat)
X_vec_cat_ed = enc.transform(X_vec_cat).toarray()
```

```python
data['count1'] = data['universal_mac_count'].shift(1)
data['count2'] = data['universal_mac_count'].shift(2)
data['count3'] = data['universal_mac_count'].shift(3)
data['count4'] = data['universal_mac_count'].shift(4)
data['count5'] = data['universal_mac_count'].shift(5)
data['count6'] = data['universal_mac_count'].shift(6)
data['count7'] = data['universal_mac_count'].shift(7)
data['count8'] = data['universal_mac_count'].shift(8)
data['count9'] = data['universal_mac_count'].shift(9)
data['count10'] = data['universal_mac_count'].shift(10)
data['count11'] = data['universal_mac_count'].shift(11)
data['count12'] = data['universal_mac_count'].shift(12)
```

```python
X = data.drop(['crowdedness','device','mac_count','packet_count','time','universal_mac_count','time_new','crowdness','hour','mini'],axis = 1)
```

```python
X.head()
```

```python
X = X.values
X = np.concatenate((X, X_vec_cat_ed), axis = 1)
```

```python
X
```

```python
X = pd.DataFrame(X)
```

```python
X
```

```python
y = data['universal_mac_count']
```

```python
X.shape
```

```python
y.shape
```

```python
y = pd.DataFrame(y)
y.index = range(len(y))
X = pd.DataFrame(X)
X_train = X.iloc[0:250]
X_test = X.iloc[250:]
y_train = y.iloc[0:250]
y_test = y.iloc[250:]
```

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
```

```python
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference = lgb_train)
params = {
    'task': 'train',
    'boosting_type': 'gbdt',  # è®¾ç½®æå‡ç±»åž‹
    'objective': 'regression', # ç›®æ ‡å‡½æ•°
    'metric': {'l2', 'auc'},  # è¯„ä¼°å‡½æ•°
    'num_leaves': 31,   # å¶å­èŠ‚ç‚¹æ•°
    'learning_rate': 0.05,  # å­¦ä¹ é€ŸçŽ‡
    'feature_fraction': 0.9, # å»ºæ ‘çš„ç‰¹å¾é€‰æ‹©æ¯”ä¾‹
    'bagging_fraction': 0.8, # å»ºæ ‘çš„æ ·æœ¬é‡‡æ ·æ¯”ä¾‹
    'bagging_freq': 5,  # k æ„å‘³ç€æ¯ k æ¬¡è¿­ä»£æ‰§è¡Œbagging
    'verbose': 1 # <0 æ˜¾ç¤ºè‡´å‘½çš„, =0 æ˜¾ç¤ºé”™è¯¯ (è­¦å‘Š), >0 æ˜¾ç¤ºä¿¡æ¯
}
```

```python
print('Start training...')
gbm = lgb.train(params,lgb_train,num_boost_round=4000,valid_sets=lgb_eval)
print('Start predicting...')
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
print('The rmse of prediction is:', mean_squared_error(y_test, y_pred) ** 0.5)
```
